```python
import pandas as pd
from selenium import webdriver
from bs4 import BeautifulSoup
import requests
```

```python
driver = webdriver.Chrome(r"C:\Users\Scp\Desktop\fliprobo\chromedriver.exe")
driver.get("https://bookpage.com/reviews")
```

```python
content = driver.page_source
soup = BeautifulSoup(content)
print(soup)
```

```python
name = soup.find_all('h4', class_ = 'italic')

Book_Name = []

for i in name:
    Book_Name.append(i.a.text)
    
len(Book_Name)
Book_Name
```

```python
ath_name = soup.find_all('p', class_ = 'sans bold')

Auther_Name = []

for i in ath_name:
    Auther_Name.append(i.text.replace('\n', ""))
    
Auther_Name
```

```python
genre = soup.find_all('p', class_ = 'genre-links hidden-phone')

Genre = []

for i in genre:
    Genre.append(i.text.replace('\n', ""))
    
Genre
```

```python
review= soup.find_all('div',class_ ="row-fluid",)
Review=[]
for i in review:
    Review.append(i.text.replace('\n',''))

Review
```

```python
df=pd.DataFrame({})
df['Book_name']= Book_Name[:5]
df['GENRE']= Genre[:5]
df['Auth_Name']= Auther_Name[:5]
df['REVIEW']= Review[:5]
```
