```python
from selenium import webdriver
from bs4 import BeautifulSoup
import pandas as pd
```

```python
driver = webdriver.Chrome(r"C:\Users\Scp\Desktop\fliprobo\chromedriver.exe")
```

```python
mov_name=[]
release=[]
mov_rate=[]
driver.get("https://www.imdb.com/chart/top?sort=ir,desc&mode=simple&page=1")
```

```python
content= driver.page_source
soup= BeautifulSoup(content)
names=soup.findAll("td", attrs={"class":"titleColumn"})
years=soup.findAll("span", attrs={"class":"secondaryInfo"})
rating=soup.findAll("td", attrs={"class":"ratingColumn imdbRating"})
print(len(names), len(years), len(rating))
```

```python
for name in names:
    mov_name.append(name.text.replace('\n',''))
```

```python
mov_name
```

```python
for year in years:
    release.append(year.text.replace('\n',''))
```

```python
release
```

```python
for i in rating:
    mov_rate.append(i.text.replace('\n',''))
```

```python
mov_rate
```

```python
len(mov_name),len(release),len(mov_rate)
```

```python
imdb=pd.DataFrame({})
imdb['Names']= mov_name[:100]
imdb['Year']= release[:100]
imdb['Rating']= mov_rate[:100]
```

```python
imdb
```

```python
imdb.to_csv("movies.csv", index=False, encoding="utf-8-sig")
```
